```python
import pandas as pd
unrate = pd.read_csv('./data/data_unrate.csv')
print(unrate.head(6))

unrate['DATE'] = pd.to_datetime(unrate['DATE'])
print(unrate.head(6))
```

```python
import matplotlib.pyplot as mplt
# matplotlib inline
# Using the different pyplot functions, we can create, customize, and display a plot.
# For example, we can use 2 functions to :
mplt.plot()
mplt.show()
```

```python
first_twelve = unrate[0:55]
#print(first_twelve)

mplt.plot(first_twelve['DATE'], first_twelve['VALUE'])
mplt.show()
```

```python
# While the y-axis looks fine, the x-axis tick labels are too close together and are unreadable
# We can rotate the x-axis tick labels by 90 degrees so they don't overlap
# We can specify degrees of rotation using a float or integer value.
mplt.plot(first_twelve['DATE'], first_twelve['VALUE'])
mplt.xticks(rotation=45)
#print(help(plt.xticks))
mplt.show()
```

```python
#xlabel(): accepts a string value, which gets set as the x-axis label.
#ylabel(): accepts a string value, which is set as the y-axis label.
#title(): accepts a string value, which is set as the plot title.

mplt.plot(first_twelve['DATE'], first_twelve['VALUE'])
mplt.xticks(rotation=36)
mplt.xlabel('Month')
mplt.ylabel('Unemployment Rate')
mplt.title('Monthly Unemployment Trends, 1948')
mplt.show()
```
