```python
import pandas as pd
import matplotlib.pyplot as mplt

unrate = pd.read_csv('./data/data_unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
#print(unrate)

first_twelve = unrate[0:55]
mplt.plot(first_twelve['DATE'], first_twelve['VALUE'])
mplt.xticks(rotation=45)
mplt.xlabel('Month')
mplt.ylabel('Unemployment Rate')
mplt.title('Monthly Unemployment Trends, 1948')
mplt.show()
```

```python
# add_subplot(first,second,index) first means number of Row,second means number of Column.
fig = mplt.figure() # 首先定义画图域
ax1 = fig.add_subplot(2,3,1) # 几行几列第几个
ax2 = fig.add_subplot(2,3,3)
ax2 = fig.add_subplot(2,3,5)
mplt.show()
```

```python
import numpy as np
# fig = mplt.figure()
fig = mplt.figure(figsize=(6,12))

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(np.random.random(5), np.random.random(5))
ax2.plot(np.arange(50)*10, np.sin(np.arange(50)))
mplt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
unrate['MONTH'] = unrate['DATE'].dt.month
# print(unrate)
fig = mplt.figure(figsize=(9,3))

mplt.plot(unrate[0:55]['DATE'], unrate[0:55]['VALUE'], c='cyan')
mplt.plot(unrate[100:155]['DATE'], unrate[100:155]['VALUE'], c=[0,0.8,1])

mplt.show()
```

```python
fig = mplt.figure(figsize=(12,6))
colors = ['red', 'cyan', 'green', 'orange', [0,1,0.5]]
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    mplt.plot(subset['MONTH'], subset['VALUE'], c=colors[i])
    
mplt.show()
```

```python
fig = mplt.figure(figsize=(12,6))
colors = ['red', 'green', 'cyan', 'orange', 'blue']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    mplt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
mplt.legend(loc='best')
# print(help(mplt.legend))
mplt.show()
```

```python
fig = mplt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    mplt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
mplt.legend(loc='upper left')
mplt.xlabel('Month, Integer')
mplt.ylabel('Unemployment Rate, Percent')
mplt.title('Monthly Unemployment Trends, 1948-1952')

mplt.show()
```
