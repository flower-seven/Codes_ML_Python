```python
import pandas as pd
reviews = pd.read_csv('./data/data_fandango_scores.csv')
#print(reviews)
cols = ['FILM', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
norm_reviews = reviews[cols]
print(norm_reviews[:1])
```

```python
# help(ax.set_xticklabels)
```

```python
import matplotlib.pyplot as mplt
import numpy as np
# The Axes.bar() method has 2 required parameters, left and height. 
# We use the left parameter to specify the x coordinates of the left sides of the bar. 
# We use the height parameter to specify the height of each bar
num_cols = ['IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = np.mean(norm_reviews[num_cols])
print(bar_heights)

bar_positions = ['IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars'] # np.arange(3) + 1
print(bar_positions)
fig, ax = mplt.subplots() # fig控制整体图的属性，ax用来画图，当画多个子图时，可用于索引，用法见help
ax.bar(bar_positions, bar_heights, 0.4) # 0.4 柱的宽度
mplt.show()
```

```python
#By default, matplotlib sets the x-axis tick labels to the integer values the bars 
#spanned on the x-axis (from 0 to 6). We only need tick labels on the x-axis where the bars are positioned. 
#We can use Axes.set_xticks() to change the positions of the ticks to [1, 2, 3, 4, 5]:

from numpy import arange
num_cols = ['IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = np.mean(norm_reviews[num_cols])
bar_positions = arange(3)+1
tick_positions = range(1,4)
fig, ax = mplt.subplots()

ax.bar(bar_positions, bar_heights, 0.4)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation=45)

ax.set_xlabel('Rating Source')
ax.set_ylabel('Average Rating')
ax.set_title('Average User Rating')
mplt.show()
```

```python
import matplotlib.pyplot as mplt
import numpy as np
num_cols = ['IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_widths = np.mean(norm_reviews[num_cols])
bar_positions = arange(3) + 1
tick_positions = range(1,4)
fig, ax = mplt.subplots()
ax.barh(bar_positions, bar_widths, 0.5)

ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols, rotation=0)
ax.set_ylabel('Rating Source')
ax.set_xlabel('Average Rating')
ax.set_title('Average User Rating')
mplt.show()
```

```python
# Let's look at a plot that can help us visualize many points.
fig, ax = mplt.subplots() # fig_scatter = mplt.figure()
ax.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['IMDB_norm'])
ax.set_xlabel('Fandango_Ratingvalue')
ax.set_ylabel('IMDB_norm') 
mplt.show()
```

```python
# Switching Axes
fig = mplt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['IMDB_norm'])
ax1.set_xlabel('Fandango')
ax1.set_ylabel('IMDB_norm')

ax2.scatter(norm_reviews['Fandango_Stars'], norm_reviews['IMDB_norm'])
ax2.set_xlabel('Fandango_Stars')
ax2.set_ylabel('IMDB_norm')
mplt.show()
```
