```python
import pandas as pd
import matplotlib.pyplot as mplt

women_degrees = pd.read_csv('./data/data_percent-bachelors-degrees-women-usa.csv')
mplt.plot(women_degrees['Year'], women_degrees['Biology'])

mplt.tick_params(bottom="off", top="off", left="off", right="off")
mplt.show()
```

```python
#100-women_degrees means men
mplt.plot(women_degrees['Year'], women_degrees['Biology'], c='blue', label='Women')
mplt.plot(women_degrees['Year'], 100-women_degrees['Biology'], c='green', label='Men')
mplt.legend(loc='best')
mplt.title('Percentage of Biology Degrees Awarded By Gender')
mplt.show()
```

```python
# Add your code here.
fig, ax = mplt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], label='Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], label='Men')

ax.tick_params(bottom="off", top="off", left="off", right="off")
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc="upper right")

mplt.show()
```

```python
fig, ax = mplt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], c='blue', label='Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], c='green', label='Men')
ax.tick_params(bottom="off", top="off", left="off", right="off")

for key,spine in ax.spines.items():
    spine.set_visible(False)
# End solution code.
ax.legend(loc='upper right')
mplt.show()
```

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = mplt.figure(figsize=(12, 12))

for sp in range(4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c='green', label='Men')
    
    mplt.legend(loc='best')

# Calling pyplot.legend() here will add the legend to the last subplot that was created.
mplt.show()
```

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = mplt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c='green', label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    
    ax.set_title(major_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    mplt.legend(loc='upper right')
    
# Calling pyplot.legend() here will add the legend to the last subplot that was created.
mplt.show()
```
