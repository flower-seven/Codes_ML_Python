```python
import pandas as pd
fd_info = pd.read_csv("./data/data_food_info.csv")
col_names = fd_info.columns.tolist()
print(col_names)
print('------------------------------------------------------------')
print(fd_info.head(3))
```

```python
print(fd_info["Iron_(mg)"])
print('------------------------------------------------------------')
iron_g = fd_info["Iron_(mg)"] / 1000
print(iron_g)
print('------------------------------------------------------------')

# Adds 100 to each value in the column and returns a Series object.
add_100 = fd_info["Iron_(mg)"] + 100
print(add_100)
print('------------------------------------------------------------')

# Subtracts 100 from each value in the column and returns a Series object.
sub_100 = fd_info["Iron_(mg)"] - 100
print(sub_100)
print('------------------------------------------------------------')

# Multiplies each value in the column by 2 and returns a Series object.
mult_2 = fd_info["Iron_(mg)"] ** 2
print(mult_2)
```

```python
#It applies the arithmetic operator to the first value in both columns, the second value in both columns, and so on
water_energy = fd_info["Water_(g)"] + fd_info["Energ_Kcal"]
water_energy = fd_info["Water_(g)"] * fd_info["Energ_Kcal"]
iron_grams = fd_info["Iron_(mg)"] / 1000
print(fd_info.shape)
fd_info["Iron_(g)"] = iron_grams
print(fd_info.shape)
```

```python
# Score=2×(Protein_(g))−0.75×(Lipid_Tot_(g))
weighted_protein = fd_info["Protein_(g)"] * 2
weighted_fat = -0.75 * fd_info["Lipid_Tot_(g)"]
initial_rating = weighted_protein + weighted_fat
```

```python
# the "Vit_A_IU" column ranges from 0 to 100000, while the "Fiber_TD_(g)" column ranges from 0 to 79
# For certain calculations, columns like "Vit_A_IU" can have a greater effect on the result, 
# due to the scale of the values
# The largest value in the "Energ_Kcal" column.
max_calories = fd_info["Energ_Kcal"].max()
print(max_calories)

# Divide the values in "Energ_Kcal" by the largest value.
normalized_calories = fd_info["Energ_Kcal"] / max_calories
print(normalized_calories)

normalized_protein = fd_info["Protein_(g)"] / fd_info["Protein_(g)"].max()
print(normalized_protein)

normalized_fat = fd_info["Lipid_Tot_(g)"] / fd_info["Lipid_Tot_(g)"].max()
print(normalized_fat)

fd_info["Normalized_Protein"] = normalized_protein
print(fd_info.shape)
fd_info["Normalized_Fat"] = normalized_fat
print(fd_info.shape)
```

```python
#By default, pandas will sort the data by the column we specify in ascending order and return a new DataFrame
# Sorts the DataFrame in-place, rather than returning a new DataFrame.
#print food_info["Sodium_(mg)"]
fd_info.sort_values("Sodium_(mg)", inplace=True) # inplace=True 在原有数据上排列，而不是生成新的列
print(fd_info["Sodium_(mg)"])

#Sorts by descending order, rather than ascending.
fd_info.sort_values("Sodium_(mg)", inplace=True, ascending=False) # ascending=False 升序=False
print(fd_info["Sodium_(mg)"])
```
