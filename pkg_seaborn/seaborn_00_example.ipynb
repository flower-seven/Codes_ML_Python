```python
data = pd.read_csv('/Users/haolei/Downloads/large_sample1.csv')
data_ses = data['SES(Z_score)'].dropna()
print(data.head(6))
#print(data_ses)
```

```python
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as mplt

fig = mplt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
sns.distplot(data_ses, bins=27, kde=True, fit=norm, hist_kws={"rwidth": 0.8, "linewidth": 3, "color": "blue"})

mplt.savefig("SES1.png")
mplt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit

hist, edges = np.histogram(data_ses, bins=27)
#print(hist)
#print(edges)
#print('---------------------')

sub_count = len(data_ses)
fig_x = edges[:-1]
#print(fig_x)

fig_y = hist/sub_count
#print(fig_y)

fig = mplt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)

width=0.13
ax.bar(fig_x, fig_y, width, ec="k",
       color=sns.color_palette("Blues",27), 
       edgecolor=sns.color_palette("Blues",27))
#ax.plot(fig_x, fig_y, c='blue')
ax.set_xlim((-3, 3))

mplt.savefig("SES2.png")
mplt.show()
```
