```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as mplt

# 生成的图直接显示在notebook中
%matplotlib inline
```

<img src="./figure/fig_01.png" alt="FAO" width="990">

```python
def sinplot(flip=1):
    x = np.linspace(0, 20, 80)
    for i in range(1, 7):
        mplt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set() #使用默认风格
sinplot()
```

### 5种主题风格 ###
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
help(sns.set_style)
```

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sinplot()
sns.despine()
```

```python
#f , ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10) # offset=10  10为图与轴线的距离
```

```python
sns.set_style("white")
sns.boxplot(data=data, palette="deep")
sns.despine(left=False,right=True,bottom=False,top=True)
```

```python
with sns.axes_style("darkgrid"): # with内采用固定风格
    mplt.subplot(211)
    sinplot()
    
mplt.subplot(212)
sinplot(-1)
```
