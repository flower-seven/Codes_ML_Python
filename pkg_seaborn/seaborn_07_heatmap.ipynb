```python
%matplotlib inline
import matplotlib.pyplot as mplt
import numpy as np; 
np.random.seed(0)
import seaborn as sns;
sns.set()
```

```python
uniform_data = np.random.rand(100,100)
print (uniform_data)
heatmap = sns.heatmap(uniform_data)
```

```python
ax = sns.heatmap(uniform_data, vmin=0.4, vmax=0.8)
```

```python
normal_data = np.random.randn(100, 100)
print (normal_data)
ax = sns.heatmap(normal_data, center=0)
```

```python
flights = sns.load_dataset("flights")
flights.head(6)
```

```python
flights = flights.pivot("month", "year", "passengers")
#print(flights)
ax = sns.heatmap(flights)
```

```python
ax = sns.heatmap(flights, annot=True, fmt="d")
```

```python
ax = sns.heatmap(flights, annot=True, fmt="d",linewidths=.8)
```

```python
ax = sns.heatmap(flights, cmap="YlGnBu", annot=True, fmt="d")
```

```python
ax = sns.heatmap(flights, cbar=False)
```
