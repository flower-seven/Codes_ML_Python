```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as mplt

import seaborn as sns
sns.set(color_codes=True)

np.random.seed(sum(map(ord, "regression")))

tips = sns.load_dataset("tips")

tips.head()
```

regplot()和lmplot()都可以绘制回归关系,推荐regplot(),因为支持参数、数据类型更多

```python
#help(sns.regplot)
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.regplot(data=tips,x="size",y="tip")
```

```python
sns.regplot(x="size", y="tip", data=tips, x_jitter=.2)
```

```python
anscombe = sns.load_dataset("anscombe")
sns.regplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 200}) # scatter_kws 点的大小
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80})
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["o", "+"], palette="Set2");
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips);
```

```python
f, ax = mplt.subplots(figsize=(5, 5))
sns.regplot(x="total_bill", y="tip", data=tips, ax=ax);
```

col_wrap：“Wrap” the column variable at this width, so that the column facets span multiple rows

size ：Height (in inches) of each facet

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips,
           col_wrap=2, size=3);
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips,
           aspect=.6);
```
